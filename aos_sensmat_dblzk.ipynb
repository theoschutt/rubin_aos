## Fitting focal Zernike basis to pupil Zernikes
Pseudocode:
```
for each sense_analysis file:
    - make dict with dof, unit_alpha, expids as keys
    - get dof, unit_alpha, list of unique exposures
    - For each exposure:
        - get rows in zkTable
        - info from each row: thx_OCS, thy_OCS, alpha, zk_OCS
        - log alpha under expid key
        - use `galsim.zernikeBasis(jmax, x, y, R_outer=1.0, R_inner=0.0)`
            - with jmax = 28?, R_outer = ? (thx_OCS, thy_OCS are in radians)
        - log the resulting focal Zernike coeffs in the expid key
    - linear fit across the exposures (i.e., N=jmax fitted (m, b) values)
    - save dict!
Plotting:
    - We will have N=jmax number of plots for each dof, each exposure contributes one point on each plot
    - For each exposure:
        - print what the dof is in title
        - get unit_alpha * alpha (this is the x value on all plots)
        - get focal_zernike vector
        - plot (unit_alpha * alpha, focal_zernike_j) across the plots
        - plot the linear fit for each focal_zernike_j
        - save plot!
```
